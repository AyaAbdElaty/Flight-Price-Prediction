```python
def duration (x):
    if type(x) == float:
        return x
    else:
        s = x.split()
        hour = s[0].split("h")
        
        if hour[0] == "5m":
            ex = hour[0].split("m")
            return int(ex[0])
        else:
            
            if len(s) == 2:
                min = s[1].split("m")
                return (int(hour[0]) * 60) + int(min[0])
            else:
                return int(hour[0]) * 60
        

    
df_train["Duration"] = df_train["Duration"].apply(duration)
df_test["Duration"] = df_test["Duration"].apply(duration)

df_train.info()


```
